```python
# imports

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
np.random.seed(0)
```

```python
n_pts = 500
X, y = datasets.make_circles(n_samples=n_pts, random_state = 123, noise=0.1, factor=0.2)
print(X) 
print(y)
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
```

```python
model = Sequential()

# Adding dense nwetwork connection
# Contains a hidden layer with 4 nodes, there is no optimal number of nodes to have.
# But having very less nodes can lead to underfitting while having too many nodes can lead to overfitting
# no.of nodes in input layer are 2
# Activation function is sigmoid function
# Ultimately, we are going to have 2 nodes in the input layer connected to 4 nodes in the hidden layer connected to 1 node in final layer.

model.add(Dense(4, input_shape=(2,), activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.compile(Adam(learning_rate = 0.01), 'binary_crossentropy', metrics=['accuracy'])
```

```python
h = model.fit(x=X, y=y, verbose = 1, batch_size = 20, epochs = 100, shuffle = "true")
# From the epochs we can observe that the accuracy reached max value at 38th epoch for this data
```

```python
plt.plot(h.history['accuracy'])
plt.legend(['accuracy'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
# From the below plot we can observe that the accuracy reached max value at before 40 epochs for this data.
```

```python
plt.plot(h.history['loss'])
plt.legend(['loss'])
plt.title('loss')
plt.xlabel('epoch')
# From the below plot we can observe that the loss keeps on decreasing till 100th epoch, eventhough accuracy is at its max at 38th epoch.
```

```python
def plot_decision_boundary(X, y, model):
    x_span = np.linspace(min(X[:,0]) - 0.25, max(X[:,0]) + 0.25)
    y_span = np.linspace(min(X[:,1]) - 0.25, max(X[:,1]) + 0.25)
    xx, yy = np.meshgrid(x_span, y_span)
    grid = np.c_[xx.ravel(), yy.ravel()]
    pred_func = model.predict(grid)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)
```

```python
plot_decision_boundary(X, y, model)
plt.scatter(X[:n_pts, 0], X[:n_pts, 1])
plt.scatter(X[n_pts:, 0], X[n_pts:, 1])
```

```python
plot_decision_boundary(X, y, model)
plt.scatter(X[:n_pts, 0], X[:n_pts, 1])
plt.scatter(X[n_pts:, 0], X[n_pts:, 1])

x = 0.1
y = 0.75
point = np.array([[x, y]])
predict = model.predict(point)
plt.plot([x], [y], marker='o', markersize=10, color="red")
print("Prediction is: ", predict)
```
